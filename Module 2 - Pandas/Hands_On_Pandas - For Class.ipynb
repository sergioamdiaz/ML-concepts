## iloc, loc and [ ] notation review:

With the <span style='color:red'>[ ]  notation </span>, items can be accessed by:
- rows: chipo[0:3] (slicing) or chipo[0] (only first row)
- columns: chipo['item_name'] or chipo.item_name
- list of columns: chipo[['item_name','item_price']]
- masking: chipo[chipo.quantity > 2]

```python
# Ex:
chipo[['item_name', 'item_price']].head()
```

With <span style='color:red'> iloc </span>, elements are accessed **by position**:

```python
chipo.iloc[ 0:5 , -1 ]  # Befores the comma is rows, after the comma is columns.
```

```python
chipo.iloc[[0,1,2,4,3],-1] # Using a list. Elements will be retrieved in the order specified in the list.
```

With <span style='color:red'> loc </span>, elements are accessed **by index**:

```python
chipo.loc[ : , 'item_name':'item_price'].head() # Slicing over columns, that's the difference with the [ ] operator.
```

### What is the average revenue amount per order?

We need first to get the revenue for each order_id, then we can get the average of those revenues.

```python
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
chipo.groupby(by=['order_id']).sum().mean()['revenue']
```
